```python
import openai

# Read the contents of the file and join the lines into a single string
text = []
with open("../../data/convictions/Det. Sam Gebbia Trial Testimony - Trial One.txt", "r", encoding="utf-8") as file:
    lines = file.readlines()
    for line in lines:
        text.append(line.rstrip('\n') + '\n')
text = ''.join(text)

# Set the maximum token limit per iteration
max_tokens_per_iteration = 1000

# Split the text into fixed token-based chunks
text_chunks = [text[i:i + max_tokens_per_iteration] for i in range(0, len(text), max_tokens_per_iteration)]

# Store the responses from each iteration
responses = []

functions = [
    {
        "name": "extract_data",
        "description": "Add the summary of a trial transcript to the database.",
        "parameters": {
            "type": "object",
            "properties": {
                "officer_name": {
                    "type": "array",
                    "items": {
                        "type": "string"
                    },
                    "description": "What are the names and roles of the police officers mentioned in this trial testimony?"
                }
            },
            "required": ["officer_name"],
        },
    }
]


# Iterate over each text chunk and send it as a message
for chunk in text_chunks:
    messages = [
        {"role": "system", "content": "You are a helpful assistant that extracts the names of law enforcement personnel, including those with ranks such as officer, detective, captain, sergeant, etc., for a database."},
        {"role": "user", "content": 'Extract a summary from the following article: ' + chunk}
    ]

    response = openai.ChatCompletion.create(
        model='gpt-3.5-turbo-0613',
        functions=functions,
        messages=messages,
        max_tokens=max_tokens_per_iteration
    )

    responses.append(response)

# Process the responses as needed
for response in responses:
    # Access and handle the relevant information from each response
    # Your code to handle the response goes here
    for choice in response.choices:
        if 'function_call' in choice['message']:
            print(choice['message']['function_call']['arguments'])
```
